```python
import pandas as pd
import plotly.express as px



df_melt = comparacao_graus.melt(
    id_vars='Grau Acadêmico',
    value_vars=['TOTAL', 'MESTRADO', 'DOUTORADO'],
    var_name='Tipo',
    value_name='Valor'
)

fig = px.bar(
    df_melt,
    x='Grau Acadêmico',
    y='Valor',
    color='Tipo',
    title='Comparação entre Graus Acadêmicos (Total, Mestrado e Doutorado)',
    labels={'Valor': 'Valor', 'Tipo': 'Tipo de Indicador'},
)

fig.update_layout(
    barmode='group',
    legend_title='Selecionar:',
)

fig.show()
```

```python
import plotly.graph_objects as go


df.columns = df.columns.str.strip()
df['TOTAL'] = pd.to_numeric(df['TOTAL'], errors='coerce')

grouped = df.groupby(['Organização Acadêmica', 'Categoria Administrativa', 'Grau Acadêmico'])['TOTAL'].sum().reset_index()


organizacoes = grouped['Organização Acadêmica'].unique()


fig = go.Figure()

for i, org in enumerate(organizacoes):
    dados_org = grouped[grouped['Organização Acadêmica'] == org]
    for categoria in dados_org['Categoria Administrativa'].unique():
        dados_cat = dados_org[dados_org['Categoria Administrativa'] == categoria]
        visibilidade = True if i == 0 else False
        fig.add_trace(go.Bar(
            x=dados_cat['Grau Acadêmico'],
            y=dados_cat['TOTAL'],
            name=categoria,
            visible=visibilidade,
        ))

n_traces_per_org = len(grouped['Categoria Administrativa'].unique())


buttons = []
for i, org in enumerate(organizacoes):
    visible = [False] * len(organizacoes) * n_traces_per_org
    for j in range(n_traces_per_org):
        trace_index = i * n_traces_per_org + j
        visible[trace_index] = True
    buttons.append(dict(label=org,
                        method='update',
                        args=[{'visible': visible},
                              {'title': f'TOTAL por Grau Acadêmico e Categoria Administrativa - {org}'}]))

fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=1.1,
        y=1.15
    )],
    barmode='group',
    title=f'TOTAL por Grau Acadêmico e Categoria Administrativa - {organizacoes[0]}',
    xaxis_title='Grau Acadêmico',
    yaxis_title='TOTAL'
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Corrigir colunas
df.columns = df.columns.str.strip()
for col in ['TOTAL', 'MESTRADO', 'DOUTORADO']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Agrupar por Organização Acadêmica, Categoria Administrativa, Grau Acadêmico
grouped = df.groupby(['Organização Acadêmica', 'Categoria Administrativa', 'Grau Acadêmico'])[['MESTRADO', 'DOUTORADO']].sum().reset_index()

# Transformar em formato longo (melt) para empilhar MESTRADO e DOUTORADO
df_long = grouped.melt(
    id_vars=['Organização Acadêmica', 'Categoria Administrativa', 'Grau Acadêmico'],
    value_vars=['MESTRADO', 'DOUTORADO'],
    var_name='Tipo',
    value_name='Valor'
)

# Organizações disponíveis
organizacoes = df_long['Organização Acadêmica'].unique()
categorias = df_long['Categoria Administrativa'].unique()
tipos = ['MESTRADO', 'DOUTORADO']

# Criar figura
fig = go.Figure()

# Adicionar uma barra para cada combinação Categoria + Tipo (para cada organização)
for i, org in enumerate(organizacoes):
    dados_org = df_long[df_long['Organização Acadêmica'] == org]
    for categoria in categorias:
        for tipo in tipos:
            dados = dados_org[(dados_org['Categoria Administrativa'] == categoria) & (dados_org['Tipo'] == tipo)]
            visibilidade = True if i == 0 else False
            fig.add_trace(go.Bar(
                x=dados['Grau Acadêmico'],
                y=dados['Valor'],
                name=f'{categoria} - {tipo}',
                visible=visibilidade
            ))

# Número de traces por organização
n_traces_per_org = len(categorias) * len(tipos)

# Dropdown buttons
buttons = []
for i, org in enumerate(organizacoes):
    visible = [False] * len(organizacoes) * n_traces_per_org
    for j in range(n_traces_per_org):
        visible[i * n_traces_per_org + j] = True
    buttons.append(dict(label=org,
                        method='update',
                        args=[{'visible': visible},
                              {'title': f'MESTRADO e DOUTORADO por Grau Acadêmico - {org}'}]))

# Layout com dropdown
fig.update_layout(
    updatemenus=[dict(
        active=0,
        buttons=buttons,
        x=1.1,
        y=1.15
    )],
    barmode='group',
    title=f'MESTRADO e DOUTORADO por Grau Acadêmico - {organizacoes[0]}',
    xaxis_title='Grau Acadêmico',
    yaxis_title='Quantidade',
    legend_title='Categoria - Tipo'
)

fig.show()
```
